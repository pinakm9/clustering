# Image Segmentation with DBSCAN

### Step - 1: Read RGB values of pixels to feed DBSCAN

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# width and height for resizing large images
width = 200
height = 200
dim = (width, height)
 
# read the image
image = cv2.imread('../../data/images/paperboats.jpg')
# resize image
image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA) 
# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# reshape the image to a 2D array of pixels and 3 color values (RGB)
pixel_values = image.reshape((-1, 3))
# convert to float
pixel_values = np.float32(pixel_values)
print(pixel_values.shape)
```

### Step - 2: Use DBSCAN to find segments

```python
from sklearn.cluster import DBSCAN
# find segments with DBSCAN
clustering = DBSCAN(eps=3, min_samples=20).fit(pixel_values)
```

### Step - 3: Reconstruct the image using the segments

```python
for label in clustering.labels_:
    if label > 0:
        members = np.where(clustering.labels_ == label)[0]
        for m in members:
            pixel_values[m] = pixel_values[members[0]]
segmented_image = pixel_values.reshape(image.shape)
# show the image
plt.imshow((segmented_image * 255).astype(np.uint8))
plt.show()
```
